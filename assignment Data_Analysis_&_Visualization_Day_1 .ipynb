## Scatter Plot

### Plotting the scatter plot

```python
x1 = 3 * np.random.rand(12, 1)
y1 = 4 + 3 *x1* np.random.randn(12, 1)
```

```python
print(x1.shape)
print(y1.shape)
```

```python
plt.scatter(x1, y1)
plt.title("Scatter plot between $X$ and $Y$")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
plt.scatter(x1, y1, c="red")
plt.title("Red colored scatter plot")
plt.xlabel("$X$ axis")
plt.ylabel("$Y$ axis")
plt.show()
```

### Changing the size

```python
plt.scatter(x1, y1, s=100)
plt.title("Increased size scatter plots")
plt.xlabel("$X axis$")
plt.ylabel("$Y axis$")
plt.show()
```

### Labeling the data points (co-ordinates)

```python
x1 = 3 * np.random.rand(10, 1)
y1 = 4 + 3 *x1* np.random.randn(10, 1)

x2 = 2 * np.random.rand(10, 1)
y2 = 2 + 4 *x1* np.random.randn(10, 1)


dataset = [(x1, y1), 
           (x2, y2)]


plt.scatter(dataset[0][0], dataset[0][1], c="red", s=20, label="From X1, Y1")
plt.scatter(dataset[1][0], dataset[1][1], c="green", s=100, label="From X2, Y2")
plt.title("Labelled data")
plt.xlabel("$X axis$")
plt.ylabel("$Y axis$")
# For enabling the labeling
plt.legend()
plt.show()
```

### Changing the shapes of "dots" (Using markers)

```python
plt.scatter(dataset[0][0], dataset[0][1], c="red", s=20, label="From X1, Y1", marker="*")
plt.scatter(dataset[1][0], dataset[1][1], c="green", s=100, label="From X2, Y2", marker="^")
plt.title("Labelled data")
plt.xlabel("$X axis$")
plt.ylabel("$Y axis$")
plt.legend()
plt.show()
```

### Changing the edge colors

```python
plt.scatter(dataset[0][0], dataset[0][1], c="red", s=100, label="From X1, Y1", marker="o", edgecolors="k")
plt.scatter(dataset[1][0], dataset[1][1], c="green", s=100, label="From X2, Y2", marker="^", edgecolors="y")
plt.title("Labelled data")
plt.xlabel("$X axis$")
plt.ylabel("$Y axis$")
plt.legend()
plt.show()
```

### Shorthand syntax for scatter plots

```python
plt.plot(dataset[0][0], dataset[0][1], "ro", label="From X1, Y1")
plt.plot(dataset[1][0], dataset[1][1], "g^", label="From X2, Y2")
plt.title("Labelled data")
plt.xlabel("$X axis$")
plt.ylabel("$Y axis$")
plt.legend()
plt.show()
```

# Line plot

```python
y1 = np.array([50, 100, 150, 200, 250])
plt.plot(y1)
plt.title("Line plot")
plt.xlabel("$X$ axis")
plt.xticks([0, 1, 2, 3, 4])
plt.yticks([50, 100, 150, 200, 250])
plt.ylabel("$Y$ axis")
plt.show()
```

```python
x1 = np.array([0, 1, 2, 3, 4])
y1 = np.array([50, 100, 150, 200, 250])
plt.plot(x1, y1)
plt.title("Line plot")
plt.xlabel("$X$ axis")
plt.ylabel("$Y$ axis")
plt.show()
```

### Changing the color of the line

```python
x1 = np.array([0, 1, 2, 3, 4])
y1 = np.array([50, 100, 150, 200, 250])
plt.plot(x1, y1, c="r")
plt.title("Line plot")
plt.xlabel("$X$ axis")
plt.ylabel("$Y$ axis")
plt.show()
```

### Changing the width of line

```python
x1 = np.array([0, 1, 2, 3, 4])
y1 = np.array([50, 100, 150, 200, 250])
plt.plot(x1, y1, linewidth=20)
plt.title("Line plot")
plt.xlabel("$X$-axis")
plt.ylabel("$Y$-axis")
plt.show()
```

### Plotting two lines on same graph

```python
x = np.array([i*2 for i in range(1, 11)])

y1 = np.array([i*2 for i in range(1, 11)])

y2 = np.array([3, 6, 9, 12, 15, 18, 21, 24, 27, 30])


dataset = [(x, y1), 
           (x, y2)]


plt.plot(dataset[0][0], dataset[0][1], c="red",label="From X1, Y1")
plt.plot(dataset[1][0], dataset[1][1], c="green", label="From X2, Y2")
plt.title("Labelled lines")
plt.xlabel("$X-axis$")
plt.ylabel("$Y-axis$")
plt.legend()
plt.show()
```
